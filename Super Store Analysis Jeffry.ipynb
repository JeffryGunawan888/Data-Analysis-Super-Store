```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Membaca DataFrame atau menggunakan DataFrame yang sudah ada
# ...

# Memilih fitur yang akan digunakan untuk clustering
features = df[['Total Sales', 'Quantity']]

# Standarisasi fitur untuk memastikan skala yang sama
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Menentukan jumlah cluster (misalnya, 3 cluster)
num_clusters = 3

# Menerapkan algoritma K-Means
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
df['Cluster'] = kmeans.fit_predict(features_scaled)

# Menampilkan hasil clustering
print(df[['Total Sales', 'Quantity', 'Cluster']])

```

```python
# Visualisasi hasil clustering
plt.scatter(df['Total Sales'], df['Quantity'], c=df['Cluster'], cmap='viridis')
plt.title('K-Means Clustering: Total Sales vs Quantity')
plt.xlabel('Total Sales')
plt.ylabel('Quantity')
plt.show()
```

```python
# total penjualan per kategori produk.

df.groupby(['Category'])[['Total Sales']].sum().reset_index().sort_values(by = ['Total Sales'],ascending = (False))
```

```python
# Filter data untuk kolom yang diperlukan
sales_by_category = df.groupby(['Category'])[['Total Sales']].sum().reset_index().sort_values(by=['Total Sales'], ascending=False)

# Tentukan warna untuk setiap kategori
colors = ['skyblue', 'lightgreen', 'lightcoral']

# Plot grafik batang dengan warna yang ditentukan
plt.figure(figsize=(10, 6))
plt.bar(sales_by_category['Category'], sales_by_category['Total Sales'], color=colors)
plt.title('Total Penjualan Berdasarkan Kategori')
plt.xlabel('Kategori')
plt.ylabel('Total Penjualan')
plt.xticks(rotation=0)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics


# Pilih fitur (X) dan target (y)
X = df[['Quantity']]  # Gantilah 'Quantity' dengan nama kolom fitur yang ingin Anda gunakan
y = df['Profit']  # Gantilah 'Profit' dengan nama kolom target

# Bagi data menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Inisialisasi model regresi linier
model = LinearRegression()

# Latih model menggunakan data latih
model.fit(X_train, y_train)

# Lakukan prediksi menggunakan data uji
y_pred = model.predict(X_test)

# Evaluasi model
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
# Visualisasi hasil prediksi
plt.scatter(X_test, y_test, color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.title('Regresi Linier - Prediksi Profit berdasarkan Quantity')
plt.xlabel('Quantity')
plt.ylabel('Profit')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Hitung Z-Score untuk kolom 'Profit'
z_scores = np.abs(stats.zscore(df['Profit']))

# Tentukan batas Z-Score sebagai threshold untuk mendeteksi anomali
threshold = 3

# Identifikasi transaksi yang memiliki Z-Score di luar threshold sebagai anomali
anomalies = df[np.abs(z_scores) > threshold]

anomalies
```

```python
# Visualisasi data dengan menyoroti anomali
plt.scatter(df.index, df['Profit'], label='Normal Transactions', color='blue')
plt.scatter(anomalies.index, anomalies['Profit'], label='Anomalies', color='red')
plt.title('Deteksi Anomali pada Profit')
plt.xlabel('Index Transaksi')
plt.ylabel('Profit')
plt.legend()
plt.show()
```
